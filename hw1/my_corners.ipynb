```python
import cv2
import numpy as np

img = cv2.imread("my_target.jpeg")
img_display = img.copy()

points = []

def click_event(event, x, y, flags, param):
    global img_display, points
    
    if event == cv2.EVENT_LBUTTONDOWN:
        points.append((x, y))
        print(f"{len(points)}: ({x}, {y})")
        
        cv2.circle(img_display, (x, y), 5, (0,0,255), -1)
        cv2.imshow("Click 20 points", img_display)
        
        if len(points) == 20:
            cv2.destroyAllWindows()

cv2.imshow("Click 20 points", img_display)
cv2.setMouseCallback("Click 20 points", click_event)

while len(points) < 20:
    cv2.waitKey(1)

v_my = np.array(points).T
print(v_my)
```
